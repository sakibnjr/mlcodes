```python
data['Your age?']=data['Your age?'].fillna(data['Your age?'].mean())
```

```python
data['Which e-learning platforms have you used?']=data['Which e-learning platforms have you used?'].fillna(data['Which e-learning platforms have you used?'].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
categorical_columns = data.select_dtypes(include=['object']).columns
print(categorical_columns)
```

```python
label_encoders = {}

for col in categorical_columns:
    label_encoders[col] = LabelEncoder()
    data[col] = label_encoders[col].fit_transform(data[col])
data
```

```python
independentFeatures = ['Your age?', 'Gender?', 'Your education level?',
       'Have you ever used e-learning platforms?',
       'Which e-learning platforms have you used?',
       'Amount of time spent daily in online classes/hour',
       'What was your attendance score on online classes?',
       'I understood the topics that are conducted online.',
       'How was your quiz score after taking online classes?',
       'Mark achieved in assignment.', 'Mark achieved in the project.',
       'Marks achieved in midterm exam.']
```

```python
x = data[independentFeatures]
x
```

```python
y = data['Tell us about your course result.']
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
select = SelectKBest(score_func=chi2)
```

```python
b_select=select.fit(x,y)
```

```python
b_select.scores_
```

```python
pd.DataFrame(b_select.scores_,columns=['Score'])
```

```python
pd.DataFrame(x.columns)
```

```python
c_score = pd.DataFrame(b_select.scores_,columns=['Score'])
```

```python
c_col = pd.DataFrame(x.columns)
```

```python
rank = pd.concat([c_col,c_score],axis=1)
```

```python
rank.nlargest(5,'Score')
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2)
```

```python
for col, encoder in label_encoders.items():
    print(f"Column: {col}")
    mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
    print(f"Mapping: {mapping}\n")
```

```python
x = data[xinput]
x
```

```python
y = data['Tell us about your course result.']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_classifier.fit(xtrain, ytrain)
```

```python
y_pred = rf_classifier.predict(xtest)
```

```python
accuracy = accuracy_score(ytest, y_pred)
print("Accuracy:", accuracy)
```
